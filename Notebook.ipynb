```python
from pathlib import Path
from markdown_pdf import MarkdownPdf, Section

def convert_md_to_pdf(md_file_path, output_pdf_path=None):
    """
    Convert a Markdown file to PDF

    Args:
        md_file_path: Path to the markdown file
        output_pdf_path: Path for the output PDF (optional)
    """
    md_path = Path(md_file_path)

    if not md_path.exists():
        print(f"Error: File not found: {md_file_path}")
        return False

    # If no output path specified, create one with same name but .pdf extension
    if output_pdf_path is None:
        output_pdf_path = md_path.with_suffix('.pdf')

    output_path = Path(output_pdf_path)

    print(f"Converting: {md_path}")
    print(f"Output to: {output_path}")

    try:
        # Read markdown content
        with open(md_path, 'r', encoding='utf-8') as f:
            md_content = f.read()

        # Convert to PDF using markdown_pdf
        pdf = MarkdownPdf(toc_level=2)
        pdf.add_section(Section(md_content, root=str(md_path.parent)))
        pdf.meta["title"] = md_path.stem
        pdf.save(str(output_path))

        print(f"✓ Successfully created PDF: {output_path}")
        return True

    except Exception as e:
        print(f"Error during conversion: {e}")
        return False

# Convert the file
md_file = "Janis Huser/Überblick/Testseite.md"
convert_md_to_pdf(md_file)
```
